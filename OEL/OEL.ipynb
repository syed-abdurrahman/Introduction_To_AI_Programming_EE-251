### Task: Analyze Building Consumption Data

#### Problem:
Using the CityLearn dataset from the 2022 competition, perform the following analysis for **Building 2**:

1. **Load the dataset** and store it in a list.
2. **Calculate the annual consumption** of energy.
3. **Calculate the monthly consumption** of energy.
4. **Calculate the daily consumption** of energy.
5. **Calculate the consumption for the four seasons**:
    - Winter: December–February
    - Spring: March–May
    - Summer: June–August
    - Autumn: September–November
6. **Calculate the daily consumption over weekdays and weekends** for the whole year.
    - Compare weekend vs weekday consumption.
7. **Visualize** and **interpret the data** in a clear and meaningful way.

---


```python
import csv

# Load the dataset
file_path = r'C:\Users\Syed Abdur Rahman\Desktop\3rd Semester\Introduction To AI\OEL\Building_2.csv'

```

```python

# Read the CSV file
with open(file_path, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    data = [row for row in reader]
```

```python

# Extracting the Equipment Electric Power values for calculations
consumption_data = [float(row['Equipment Electric Power [kWh]']) for row in data]
```

```python

# Calculate annual consumption
annual_consumption = sum(consumption_data)
```

```python

# Calculate monthly consumption
monthly_consumption = {}
for row in data:
    month = int(float(row['Month']))
    power = float(row['Equipment Electric Power [kWh]'])
    if month in monthly_consumption:
        monthly_consumption[month] += power
    else:
        monthly_consumption[month] = power
```

```python

# Calculate daily consumption
daily_consumption = {}
for row in data:
    day = int(float(row['Day Type']))
    power = float(row['Equipment Electric Power [kWh]'])
    if day in daily_consumption:
        daily_consumption[day] += power
    else:
        daily_consumption[day] = power
```

```python

# Seasonal consumption
seasons = {
    "Winter": [12, 1, 2],
    "Spring": [3, 4, 5],
    "Summer": [6, 7, 8],
    "Autumn": [9, 10, 11]
}

seasonal_consumption = {}
for season, months in seasons.items():
    seasonal_consumption[season] = sum(monthly_consumption[month] for month in months)
```

```python

# Weekday vs Weekend consumption
weekday_consumption = 0
weekend_consumption = 0
weekday_count = 0
weekend_count = 0

for row in data:
    day_type = int(float(row['Day Type']))
    power = float(row['Equipment Electric Power [kWh]'])
    
    if day_type in [1, 2, 3, 4, 5]:  # Weekdays
        weekday_consumption += power
        weekday_count += 1
    elif day_type in [6, 7]:  # Weekends
        weekend_consumption += power
        weekend_count += 1
```

```python

# Average consumption for weekdays and weekends
average_weekday_consumption = weekday_consumption / weekday_count if weekday_count else 0
average_weekend_consumption = weekend_consumption / weekend_count if weekend_count else 0
```

```python

# Output the results
print(f"Annual Consumption: {annual_consumption:.2f} kWh\n")
print("Monthly Consumption (kWh):")
for month, consumption in monthly_consumption.items():
    print(f"Month {month}: {consumption:.2f} kWh")
 
```

```python
   
print("\nSeasonal Consumption (kWh):")
for season, consumption in seasonal_consumption.items():
    print(f"{season}: {consumption:.2f} kWh")
```

```python
    
print(f"\nAverage Weekday Consumption: {average_weekday_consumption:.2f} kWh")
print(f"Average Weekend Consumption: {average_weekend_consumption:.2f} kWh")
```
