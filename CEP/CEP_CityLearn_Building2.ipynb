### Question 1
Load the **building_2 CSV file** and the **price CSV file**.  
Save the **“Equipment Electric Power [kWh]”** as the building’s electricity consumption in a list named **non-shiftable load**.  
Calculate its **net sum**, which represents the building’s **net electricity consumption**.


```python

non_shiftable_load = []

with open("building_2.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        non_shiftable_load.append(float(row["Equipment Electric Power [kWh]"]))  #it take the equipment electric data from building_2.csv

net_consumption_building = sum(non_shiftable_load)   #here we use the SUM built in function from addind all the values
print("Net electricity consumption (without PV):", net_consumption_building, "kWh")

```

### Question 2
Store the **“Solar Generation [W/kW]”** in a list as **PV generation per kW** of the installed system.  
Convert it into actual solar generation using the formula:

PVgen = (4 × Solar Generation [W/kW]) / 1000


After conversion, store the solar generation in a list named **Net solar generation of building 2**.


```python

pv_generation_per_kw = []          #  new list for PV Generation
net_solar_generation_building_2 = []

with open("building_2.csv", "r") as file:    # it open the file
    reader = csv.DictReader(file)
    for row in reader:
        pv = float(row["Solar Generation [W/kW]"])
        pv_generation_per_kw.append(pv)
        net_solar_generation_building_2.append((4 * pv) / 1000)    # formula in the question

print("Total solar generation:", sum(net_solar_generation_building_2), "kWh")
```

### Question 3
Store the electricity **price** in a list named **price**.  
Multiply the price by the electricity consumption **hour by hour** and store the result in a list named **bill without PV**.  
Calculate and display the **net sum of the bill without PV**.


```python
import csv

Price = []

with open('pricing.csv') as file:
    reader = csv.reader(file)
    for index, R_1 in enumerate(reader):    
        if index == 0:  # skip header row
            continue
        Price.append(float(R_1[1]))  # append 1th column
```

```python

bill_without_pv = []

for i in range(len(non_shiftable_load)):    
    bill_without_pv.append(non_shiftable_load[i] * Price[i])

print("Total bill without PV:", sum(bill_without_pv))
```

### Question 4
Subtract the solar generation from the electricity consumption.  
Since this results in both **positive and negative values**, store the result in a list named **Net consumption**.


```python

net_consumption = []

for i in range(len(non_shiftable_load)):
    net_consumption.append(non_shiftable_load[i] - net_solar_generation_building_2[i])
```

### Question 5
Split the **Net consumption** into:
- A list of **positive values** named **Net electricity consumption with PV**
- A list of **negative values**, representing **wasted (surplus) energy** that can be exported to the grid or stored in a battery  

Calculate and display the **net sum** of both electricity consumption with PV and wasted energy.


```python

net_electricity_consumption_with_pv = []
wasted_energy = []

for value in net_consumption:
    if value > 0:     
        net_electricity_consumption_with_pv.append(value)
    else:
        wasted_energy.append(value)

print("Net electricity consumption with PV:", sum(net_electricity_consumption_with_pv), "kWh")
print("Total wasted (surplus) energy:", sum(wasted_energy), "kWh")
```
